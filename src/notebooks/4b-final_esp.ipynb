```python
# Lettura dataset
prov11 = pd.read_excel("../../data/raw/esp/AE_2010_13_01_02_03 (1).xls",skiprows=4)

# Ispezioniamo le prime righe del dataset
prov11.head()
display(prov11)
```

```python
# Select the relevant columns for 'Superficie Total' and 'Producción de grano'
df_selected11 = prov11[['Provincias y Comunidades Autónomas',  'Unnamed: 3', 'Producción']]

# Rename columns for clarity
df_selected11.columns = ['Provincia', 'Superficie Total', 'Producción Grano']

# Clean any rows where the 'Provincia' field is NaN (header rows or empty rows)
df_selected11 = df_selected.dropna(subset=['Provincia'])

# Aggiungi una colonna "Anno" con valore 2006 per tutte le righe
df_selected11['Anno'] = 2009


# Inspect the cleaned data with relevant columns
df_selected11.head()
```

```python
# Supponiamo che i tuoi DataFrame si chiamino df_selected1, df_selected2, ..., df_selectedN
# Li inseriamo in una lista
dfs = [df_selected,df_selected1, df_selected2, df_selected3, df_selected4, df_selected5, df_selected6, df_selected7, df_selected8, df_selected9, df_selected10, df_selected11]

# Usiamo pd.concat() per unire tutti i DataFrame in verticale
df_selectedFinal = pd.concat(dfs, ignore_index=True)

df_selectedFinal = df_selectedFinal.dropna() 

# Visualizza i primi dati del DataFrame unito

display(df_selectedFinal)
print(df_selectedFinal.dtypes)
```

```python
# Converte le colonne in numerico, sostituendo eventuali errori (valori non numerici) con NaN
df_selectedFinal['Superficie Total'] = pd.to_numeric(df_selectedFinal['Superficie Total'], errors='coerce')
df_selectedFinal['Producción Grano'] = pd.to_numeric(df_selectedFinal['Producción Grano'], errors='coerce')
```

```python
# Raggruppa i dati in base alle colonne ['comune', 'provincia', 'alt', 'lat', 'lon'].
# Questo significa che per ogni combinazione di questi valori, verranno calcolate le statistiche di aggregazione.
# Per ogni comune, vengono calcolate:
# superficie_media: la media della colonna 'superficie totale - ettari'.
# produzione_media: la media della colonna 'produzione totale - quintali'.
# superficie_std: la deviazione standard della colonna 'superficie totale - ettari'.
# produzione_std: la deviazione standard della colonna 'produzione totale - quintali'.


# Calcolo delle statistiche di aggregazione raggruppando per Provincia
features = df_selectedFinal.groupby('Provincia').agg(
    superficie_media=('Superficie Total', 'mean'),
    produzione_media=('Producción Grano', 'mean'),
    superficie_std=('Superficie Total', 'std'),
    produzione_std=('Producción Grano', 'std')
).reset_index()

# Visualizza i primi risultati
display(features)
```

```python
# Converti i nomi delle province in minuscolo
features['Provincia'] = features['Provincia'].str.title()

features['produzione_media'] = features['produzione_media'] * 10

# Visualizza i primi risultati per verificare
display(features)
```

```python
# Lettura dataset
final = pd.read_excel("../../data/raw/esp/dati_esp.xlsx")

# Ispezioniamo le prime righe del dataset
final.head()
display(final)
```

```python
# Rinomina le colonne 'Unnamed: 1' e altre
final.rename(columns={'Unnamed: 1': 'Provincia', 'Alt': 'Altitudine', 'Lat': 'Latitudine', 'Long': 'Longitudine', 'prec media': 'Precipitazione Media', 'temp media': 'Temperatura Media'}, inplace=True)

# Elimina la colonna 'Unnamed: 0'
final.drop(columns=['Unnamed: 0'], inplace=True)

# Visualizza i primi risultati per verificare
display(final)
```

```python
# Unione dei due dataset sulla colonna 'Provincia'
df_not_ord = pd.merge(features,final, on='Provincia', how='inner')

# Visualizza i primi risultati per verificare
display(df_not_ord)
```

```python
df_not_ord.columns = df_not_ord.columns.str.strip()

df_not_ord.rename(columns={'Provincia': 'provincia', 'Altitudine': 'alt', 'Lat': 'lat', 'Longitudine': 'lon', 'Precipitazione Media': 'prec_media', 'Temperatura Media': 'temp_media'}, inplace=True)

df_ord = df_not_ord[['provincia', 'alt', 'lat', 'lon', 'superficie_media', 'produzione_media', 'prec_media', 'temp_media', 'superficie_std', 'produzione_std']]
```

```python
df_ord['alt'] = df_ord['alt'].astype('float64')
df_ord['lat'] = pd.to_numeric(df_ord['lat'], errors='coerce')
df_ord['lon'] = pd.to_numeric(df_ord['lon'], errors='coerce')

print(df_ord.dtypes)
```

```python
final = df_ord.dropna()

display(final)

final.to_csv("../../data/processed/features_esp.csv")
```
